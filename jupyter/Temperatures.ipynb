```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
url = 'https://raw.githubusercontent.com/stefmolin/Hands-On-Data-Analysis-with-Pandas-2nd-edition/master/ch_03/data/nyc_temperatures.csv'
```

```python
df = pd.read_csv(url, sep=",")
```

```python
weatherset = df
```

```python
weatherset.columns
```

```python
# rename column
weatherset.rename(columns={'value':'temp'}, inplace=True)
```

```python
# checking data
weatherset.head()
```

```python
# checking type of data
weatherset.dtypes
```

```python
'''
As you see, date has timestamp, different for analysis.
Okay, convert it to date time.
'''
df.loc[:, 'date'] = pd.to_datetime(weatherset.date)
```

```python
'''
Order, sorting data with temp
but order by date in valid with column max
'''
weatherset.sort_values(by='temp', ascending=False).head(10)
```

```python
weatherset[weatherset.datatype == 'TMAX'].sort_values(by=['date','temp'], ascending=False).head(10)
```
